```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Weather.csv')
df
```

```python
df.info()
```

```python
duplicated_rows = df.duplicated(subset=['Date/Time'])
duplicated_rows
```

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True),cmap='Blues', annot=True)
plt.figure(figsize=((3,3)))
plt.show
```

```python
#scatter plot between Temperature and Humidity

plt.figure(figsize=(10, 6))
plt.scatter(df['Temp_C'], df['Rel Hum_%'], alpha=0.3)
plt.xlabel('Temperature (C)')
plt.ylabel('Relative Humidity (%)')
plt.title('Temperature vs Relative Humidity')
plt.show()
```

```python
plt.figure(figsize=(30, 15))
sns.boxplot(x='Weather', y='Temp_C', data=df)
plt.xlabel('Weather Condition')
plt.ylabel('Temperature (C)')
plt.title('Temperature Distribution Across Weather Conditions')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```
